```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/mental_health_data.csv")
print(df)
```

```python
# Shape of dataset
df.shape
```

```python
# Checking Data Types
df.info()
```

```python
#Check Missing Values
df.isnull().sum()
```

```python
# Remove Unnecessary Columns
df_clean = df.drop(columns=[
"Low CI",
"High CI",
"Confidence Interval",
"Quartile Range"
])
```

```python
# Handle Missing Values
df_clean = df_clean.dropna(subset=["Value"])

df_clean.isnull().sum()
```

```python
# Convert Date Columns
df_clean["Time Period Start Date"] = pd.to_datetime(df_clean["Time Period Start Date"], format='mixed')
df_clean["Time Period End Date"] = pd.to_datetime(df_clean["Time Period End Date"], format='mixed')
```

```python
# Filter Useful Segmentation Groups
df_filtered = df_clean[
df_clean["Group"].isin([
"By Age",
"By Sex",
"By Race/Hispanic ethnicity",
"By State"
])
]
```

```python
# Checking final dataset.
df_filtered.shape

df_filtered.head()
```

```python
#Save Clean Dataset
df_filtered.to_csv("/content/mental_health_data - Copy.csv", index=False)

print("Cleaned CSV file saved successfully ✅")
```
